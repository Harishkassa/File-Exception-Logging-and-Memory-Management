24. How do you raise an exception manually in Python

```python
class MyExceptionError(Exception):
  pass

def test(n):
  if n < 0:
    raise MyExceptionError("Number should be positive")


try:
  test(-5)
except MyExceptionError as e:
  print(e)
```

21. Write a Python program that reads a file and prints the number of occurrences of a specific word

```python
target_word = "Python"

with open("myfile.txt", "r") as f:
    content = f.read().lower()  # convert to lowercase once
    count = content.count(target_word.lower())  # use str.count()

print(f"The word '{target_word}' occurs {count} times in the file.")
```

22. How can you check if a file is empty before attempting to read its contents

```python
import os

if os.path.exists("myfile.txt"):
  if os.stat("myfile.txt").st_size == 0:
    print("file is empty")

  else:
    with open("myfile.txt","r") as f:
      content = f.read()
      print(content)

else:
  print("file not found")
```
